```python
12.#what are the Top 10 countries with the highest average AQI value
```

```python
#Top 10 countries with the highest average AQI value
aqi_avg = data.groupby('Country')['AQI Value'].mean().reset_index()
highaqi_avg = aqi_avg.sort_values('AQI Value', ascending=False)
print("Top 10 Countries with High AQI Values","\n",highaqi_avg.head(10))


```

```python
13.what are the top  10 countries haveing lowest AQI
```

```python
#Top 10 countries with the lowest average AQI value
aqi_avg = data.groupby('Country')['AQI Value'].mean().reset_index()
lowaqi_avg = aqi_avg.sort_values('AQI Value', ascending=True)
print("Top 10 Countries with Low AQI Values","\n",lowaqi_avg.head(10))
```

```python
14.Comparison of AQI Categories by Pollutant
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=data, x='AQI Category', y='AQI Value')
plt.title('Distribution of AQI Categories')
plt.xlabel('AQI Category')
plt.ylabel('AQI Value')
plt.show()
```

```python
15.make a corr_matrix
```

```python
corr_matrix = data[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Pollutant AQI Values')
plt.show()
```
